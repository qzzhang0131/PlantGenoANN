```python
import os
import pyBigWig
import numpy as np
import matplotlib.pyplot as plt
```

```python
output_dir = "./example/"
species = "Arabidopsis_lyrata"
chrom_name = "8"
start_pos = 490000
end_pos = 500000
```

```python
features = ["Gene", "Exon", "CDS"]

colors = {
    "Gene_+": "#e41a1c",   
    "Gene_-": "#377eb8",
    "Exon_+": "#4daf4a",
    "Exon_-": "#984ea3",
    "CDS_+":  "#ff7f00",
    "CDS_-":  "#a65628"
}

fig, axes = plt.subplots(6, 1, figsize=(14, 12), sharex=True)
x_axis = np.arange(start_pos, end_pos)

for i, feature in enumerate(features):
    feat_str = feature.lower() if feature != "CDS" else "CDS"
    pos_file = os.path.join(base_dir, f"{species}_{feat_str}_in_positive_strand.bw")
    neg_file = os.path.join(base_dir, f"{species}_{feat_str}_in_negative_strand.bw")
    
    ax_pos = axes[i * 2]
    bw_pos = pyBigWig.open(pos_file)
    y_pos = np.nan_to_num(bw_pos.values(chrom_name, start_pos, end_pos))
    bw_pos.close()
    
    ax_pos.plot(x_axis, y_pos, color=colors[f"{feature}_+"], linewidth=1.2, label=f"{feature} (+)")
    ax_pos.set_ylim(0, 1.05)
    ax_pos.set_ylabel("Prob")
    ax_pos.legend(loc="upper right")
    ax_pos.grid(axis='x', linestyle='--', alpha=0.5)
    
    ax_neg = axes[i * 2 + 1]
    bw_neg = pyBigWig.open(neg_file)
    y_neg = np.nan_to_num(bw_neg.values(chrom_name, start_pos, end_pos))
    bw_neg.close()

    ax_neg.plot(x_axis, y_neg, color=colors[f"{feature}_-"], linewidth=1, label=f"{feature} (-)")
    ax_neg.set_ylim(0, 1.05)
    ax_neg.set_ylabel("Prob")
    ax_neg.legend(loc="upper right")
    ax_neg.grid(axis='x', linestyle='--', alpha=0.5)

axes[-1].set_xlabel(f"Genomic Position on {chrom_name} (bp)", fontsize=12)
plt.xlim(start_pos, end_pos)
plt.tight_layout()

plt.show()
```
